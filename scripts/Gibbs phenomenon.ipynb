```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np, scipy as sp

# For√ßar o uso de LaTeX
from matplotlib import rc
##rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 

def sum_of_sine(N, x):
    """ Calculates \sum_{i = 1}{N} sin([2*i + 1] x)/(2i+1) for given N, x."""
    
    pts = np.arange(1, N+1, 2) # Odd points
    return sum(np.sin(pts * x)/pts)
        
```

```python
plt.figure(1, figsize=(12,9))
plt.grid()
xvals = np.arange(0, 2*np.pi, 0.001)
#plt.axvline(np.pi)
vals = [20, 200, 500]
leg = []

for pos, N in enumerate(vals):
    plt.subplot(311 + pos)
    plt.title(u"Fen\^omeno de Gibbs, N = {}".format(N), fontsize=16)
    plt.tight_layout()
    plot(xvals, [sum_of_sine(N, x) for x in xvals], 'k', linewidth=1.5)
    plt.xlim([0, 2*np.pi])

    #plt.legend(["N = {}".format(N)])
#plt.legend(leg)
plt.savefig('gibbs.eps')
```
