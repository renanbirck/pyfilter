```python
%pylab inline
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

# Forçar o uso de LaTeX
from matplotlib import rc
##rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 

N = 6    # Ordem dos filtros
Wn = 1   # Freq. -3 dB = 1 rad/s
Rp = 1   # Ripple na banda de passagem (dB)
Rs = 80  # Atenuacao na banda de parada (dB)

buttfilt = signal.butter(N, Wn, 'low', analog=True)
cheb1filt = signal.cheby1(N, Rp, Wn, 'low', analog=True)
cheb2filt = signal.cheby2(N, Rs, Wn, 'low', analog=True)
bessfilt = signal.bessel(N, Wn, 'low', analog=True)
ellipfilt = signal.ellip(N, Rp, Rs, Wn, 'low', analog=True)

freqs_range = np.linspace(1e-1, 1e1, 1000)
butt_w, butt_h = signal.freqs(*buttfilt, worN=freqs_range)
cheb1_w, cheb1_h = signal.freqs(*cheb1filt, worN=freqs_range)
cheb2_w, cheb2_h = signal.freqs(*cheb2filt, worN=freqs_range)
bess_w, bess_h = signal.freqs(*bessfilt, worN=freqs_range)
ellip_w, ellip_h = signal.freqs(*ellipfilt, worN=freqs_range)

fig = plt.figure(1, figsize=(12,9))

p1, = plt.plot(butt_w, 20 * np.log10(np.abs(butt_h)), 'k-')
p2, = plt.plot(cheb1_w, 20 * np.log10(np.abs(cheb1_h)), 'k--')
plt.plot(cheb2_w, 20 * np.log10(np.abs(cheb2_h)), 'k.')
plt.plot(bess_w,  20 * np.log10(np.abs(bess_h)), 'k.-')
plt.plot(ellip_w,  20 * np.log10(np.abs(ellip_h)), 'k-,')
ax = plt.gca()

plt.xscale('log')
plt.grid(markevery=0.01)
plt.margins(0, 0.1)
plt.title(u"Resposta em frequ\^encia comparada para $\omega_c = {}$ rad/s ".format(Wn), fontsize=16)
plt.xlabel(u"Frequ\^encia (rad/s)", fontsize=19)
plt.ylabel("Ganho (dB)", fontsize=19)
plt.legend(["Butterworth", "Chebyshev I", "Chebyshev II", "Bessel", u"El\\'iptico"], fontsize=14, loc='best')
plt.axvline(1, color='green') # Marcar freq. corte
plt.savefig('compare_frequencies.svg')

plt.show()
```
