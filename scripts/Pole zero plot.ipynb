```python
%matplotlib inline
from scipy import signal
import matplotlib.pyplot as plt, matplotlib
import numpy as np
import scipy as sp

# Forçar o uso de LaTeX
from matplotlib import rc
##rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18) 
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10 # gráficos quadrados
```

```python
[butter_zeros, butter_poles] = signal.butter(10, 1, analog=True, output='ba')
butter_roots = np.roots(butter_poles)
[cheby1_zeros, cheby1_poles] = signal.cheby1(10, 1, 1, analog=True, output='ba')
cheby1_roots = np.roots(cheby1_poles)
[cheby2_zeros, cheby2_poles] = signal.cheby2(10, 80, 1, analog=True, output='ba')
cheby2_roots = np.roots(cheby2_poles)
[ell_zeros, ell_poles] = signal.ellip(10, 1, 80, 1, analog=True, output='ba')
ell_roots = np.roots(ell_poles)
[bessel_zeros, bessel_poles] = signal.bessel(10, 1, analog=True, output='ba')
bessel_roots = np.roots(bessel_poles)
```

```python
fig, ((ax_butter, ax_bessel), (ax_cheby1, ax_cheby2), (ax_ellip, ax_empty)) = plt.subplots(nrows=3, ncols=2, figsize=(10, 15))
ax_butter.scatter(butter_roots.real, butter_roots.imag)
ax_butter.grid()
ax_butter.set_aspect(1)
ax_butter.set_xlim(-1.5, 1.5)
ax_butter.set_ylim(-1.5, 1.5)
ax_butter.set_title('Butterworth')

ax_bessel.scatter(bessel_roots.real, bessel_roots.imag)
ax_bessel.grid()
ax_bessel.set_aspect(1)
ax_bessel.set_xlim(-1.5, 1.5)
ax_bessel.set_ylim(-1.5, 1.5)
ax_bessel.set_title('Bessel')

ax_cheby1.scatter(cheby1_roots.real, cheby1_roots.imag)
ax_cheby1.grid()
ax_cheby1.set_aspect(1)
ax_cheby1.set_xlim(-1.5, 1.5)
ax_cheby1.set_ylim(-1.5, 1.5)
ax_cheby1.set_title('Chebyshev I')

ax_cheby2.scatter(cheby2_roots.real, cheby2_roots.imag)
ax_cheby2.grid()
ax_cheby2.set_aspect(1)
ax_cheby2.set_xlim(-1.5, 1.5)
ax_cheby2.set_ylim(-1.5, 1.5)
ax_cheby2.set_title('Chebyshev II')

ax_ellip.scatter(ell_roots.real, ell_roots.imag)
ax_ellip.grid()
ax_ellip.set_aspect(1)
ax_ellip.set_xlim(-1.5, 1.5)
ax_ellip.set_ylim(-1.5, 1.5)
ax_ellip.set_title(u"El\\'iptico")


for ax in fig.get_axes():
    ax.set_xlabel('Re(z)', fontsize=15)
    ax.set_ylabel('Im(z)', fontsize=15)
fig.delaxes(ax_empty)
fig.tight_layout()

fig.savefig('../drawings/filter_comparison.svg')
```
